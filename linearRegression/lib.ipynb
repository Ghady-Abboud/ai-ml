```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("data.csv")
print(df.head())
data_X, data_Y = df[["study_time"]], df[["score"]]

data_X_train = data_X[:-20]
data_X_test = data_X[-20:]

data_Y_train = data_Y[:-20]
data_Y_test = data_Y[-20:]
```

```python
regr = LinearRegression()
regr.fit(data_X_train, data_Y_train)
data_Y_pred = regr.predict(data_X_test)
```

```python
coef = regr.coef_
intercept = regr.intercept_
mse = mean_squared_error(data_Y_test, data_Y_pred)
r2 = r2_score(data_Y_test, data_Y_pred)
print(f"Coef: {coef}\n")
print(f"Intercept: {intercept}\n")
print(f"MSE: {mse}\n")
print(f"R2 Score: {r2}\n")
```

```python
plt.scatter(df["study_time"], df["score"])
plt.xlabel("Study Time")
plt.ylabel("Score")
plt.plot(df["study_time"], coef[0][0] * df["study_time"] + intercept, color="red")
plt.show()
```
